```python
from matplotlib import pyplot as plt
import numpy as np
import pandas
from scipy import stats
```

```python
#read hail pixel data
hailpx_major_px = np.load('all_major.npy')
hailpx_minor_px = np.load('all_minor.npy')
#read hail pad data
df = pandas.read_csv('pad144-2.csv', header=None)
as_dict = df.to_dict(orient='list')
pad_major_mm = np.array(as_dict[0])
pad_minor_mm = np.array(as_dict[1])

count_pad       = len(pad_major_mm)
count_hailpx    = len(hailpx_major_px)
hail_sz_bins    = np.arange(20,50,2.5)
axis_ratio_bins = np.arange(0.5,1.05,0.05)
```

```python
def calc_bin(bin_left_vec, x_data, y_data):
    bin_width = bin_left_vec[1] - bin_left_vec[0]
    bin_mean_vec = []
    bin_std_vec  = []
    for i, bin_left in enumerate(bin_left_vec):
        bin_right = bin_left+bin_width
        bin_mask = np.logical_and(x_data>=bin_left, x_data<bin_right)
        bin_data = y_data[bin_mask]
        bin_mean_vec.append(np.nanmean(bin_data))
        bin_std_vec.append(np.nanstd(bin_data))
    print(bin_std_vec)
    return bin_mean_vec, bin_std_vec
```

```python
#apply scaling to hailpixel data
pixel_scaling = 2.54 #mm/pixel
hailpx_major_mm = hailpx_major_px*pixel_scaling
hailpx_minor_mm = hailpx_minor_px*pixel_scaling

# init plot
fig = plt.figure(facecolor='white',figsize=[17, 10])

plt.subplot(2, 3, 1)
plt.hist(hailpx_major_mm, density=True, color = 'blue', edgecolor = 'black', bins=hail_sz_bins)
#fit with beta distribtion
fit_data = stats.beta.fit(hailpx_major_mm)
fit_y = stats.beta(fit_data[0], fit_data[1], fit_data[2], fit_data[3])
print(fit_y)
plt.plot(hail_sz_vec+1.25, fit_y.pdf(hail_sz_vec+1.25), 'k-')
#
plt.ylabel('probability density',fontsize=14)
plt.xlim(15, 50)

plt.subplot(2, 3, 2)
plt.hist(hailpx_minor_mm/hailpx_major_mm, density=True, color = 'blue', edgecolor = 'black', bins=axis_ratio_bins)
plt.ylabel('probability density',fontsize=14)
plt.xlim(0.5, 1.0)

plt.subplot(2, 3, 3)
plt.plot(hailpx_major_mm, hailpx_minor_mm/hailpx_major_mm,'b.')
plt.ylabel('axis ratio',fontsize=14)
#fit with 5mm bins
fit_bins = np.arange(20,50,5)
bin_mean_vec, bin_std_vec = calc_bin(fit_bins, hailpx_major_mm, hailpx_minor_mm/hailpx_major_mm)
plt.errorbar(fit_bins+2.5, bin_mean_vec, bin_std_vec, ecolor='k', fmt='k-*', zorder=10)

plt.xlim(15, 50)
plt.ylim(0.5, 1.0)

plt.subplot(2, 3, 4)
plt.hist(pad_major_mm, density=True, color = 'red', edgecolor = 'black', bins=np.arange(20,50,2.5))
plt.xlabel('major axis length (mm)',fontsize=14)
plt.ylabel('probability density',fontsize=14)
plt.xlim(15, 50)

plt.subplot(2, 3, 5)
plt.hist(pad_minor_mm/pad_major_mm, density=True, color = 'red', edgecolor = 'black', bins=np.arange(0.5,1.05,0.05))
plt.xlabel('axis ratio',fontsize=14)
plt.ylabel('probability density',fontsize=14)
plt.xlim(0.5, 1.0)

plt.subplot(2, 3, 6)
plt.plot(pad_major_mm, pad_minor_mm/pad_major_mm,'r.')
plt.ylabel('axis ratio',fontsize=14)
plt.xlabel('major axis length (mm)',fontsize=14)
plt.xlim(15, 50)
plt.ylim(0.5, 1.0)

plt.savefig('compare_hist', dpi=150)
```
