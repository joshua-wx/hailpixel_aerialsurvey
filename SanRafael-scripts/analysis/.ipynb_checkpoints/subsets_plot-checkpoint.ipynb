```python
json_ffn     = '/g/data1a/kl02/jss548/hail-research/HailPixel/subsets/hailpixel_20181126_final_hq_complete.json'
ref_ffn      = '/g/data1a/kl02/jss548/hail-research/HailPixel/subsets/index.csv'
```

```python
#read reference file
df = pd.read_csv(ref_ffn, header=0)
ref_dict = df.to_dict(orient='list')
```

```python
#build a list of subset x and y centroid coordainted for subsets
final_subset_id_list = []
padding  = 50
centroid_offset = 250

#open json index file
with open(json_ffn) as json_fh:  
    data = json.load(json_fh)

#load keys from file
json_keys = list(data.keys())[0:]

for key_idx in range(0, len(json_keys)):
    #extract image json
    image_json = data[json_keys[key_idx]]
    #loop through regions in annotation file
    region_json = image_json['regions']
    
    #find reference coordinates for subset
    subset_idx = int(json_keys[key_idx][0:4])
    ref_idx    = ref_dict['idx'].index(int(subset_idx))
    
    #add subset coordinates
    final_subset_id_list.append(ref_idx)
```

```python
# plot image and subset regions

padding   = 50
subset_sz = 500

fig = plt.figure(facecolor='white',figsize=[20,10])
plt.imshow(im_padded)
plt.axis('off')

for i, subset_id in enumerate(ref_dict['idx']):
    
    x0 = ref_dict['x0'][i] + padding
    y0 = ref_dict['y0'][i] + padding
    
    square_x = [x0, x0+subset_sz, x0+subset_sz, x0, x0] 
    square_y = [y0, y0, y0+subset_sz, y0+subset_sz, y0]
    
    if subset_id in final_subset_id_list:
        plt.plot(square_x, square_y, 'k-')
    #else:
        #plt.plot(square_x, square_y, 'w-')
    
plt.savefig('swath_subsets_figure.png', dpi=150)

# plt.ylim((0,17500))
# plt.xlim((0,17500))
```
